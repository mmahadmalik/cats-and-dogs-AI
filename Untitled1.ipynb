```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the DataFrame with irradiance data from the JSON file
df_latitude = pd.read_json("/content/drive/MyDrive/2. Ted Nash/All codes /Code for finding the angle value from the latitude value/JSON file Testing/Lattitude.json")


# Calculate δ for each day for whole year
n_values = np.arange(1, 366)
delta_values = 23.45 * np.sin(2 * np.pi * (n_values - 81) / 365)

# Create a DataFrame with δ values for all 365 days
df_delta = pd.DataFrame({'Day': np.repeat(n_values, len(df_latitude)), 'Delta': np.tile(delta_values, len(df_latitude))})

# Merge δ values with the original DataFrame
# Merge δ values with the original DataFrame
df_latitude = pd.concat([df_latitude] * 365, ignore_index=True)
df_latitude = pd.concat([df_latitude, pd.DataFrame(df_delta['Delta'])], axis=1)

# Calculate β for each latitude and day
for column in df_latitude.columns:
    if 'Lattitude' in column:
        Beta_values = (90 - df_latitude[column] + df_latitude['Delta'])
        df_latitude[f'Beta_{column.split(" ")[1]}'] = Beta_values

# Calculate Tilt for each latitude and day
for column in df_latitude.columns:
    if 'Lattitude' in column:
        Tilt_values = 90 - (90 - df_latitude[column] + df_latitude['Delta'])
        df_latitude[f'Tilt_{column.split(" ")[1]}'] = Tilt_values
        # Save the updated DataFrame to a new Excel file

#Now lets make the graph of this angles for the customer



# Calculate Best Tilt for each latitude and day
#for column in df_latitude.columns:
#    if 'Lattitude' in column:
#        Best_Tilt1 = 90 - (90 - df_latitude[column] + df_latitude['Delta'])
#        df_latitude[f'Best_{column.split(" ")[1]}'] = Best_Tilt1
#        Best_Tilt = df_latitude[f'Best_{column.split(" ")[1]}'] / 365


# Save the updated DataFrame to a new Excel file
df_latitude.to_excel('/content/drive/MyDrive/2. Ted Nash/All codes /Code for finding the angle value from the latitude value/JSON file Testing/output_Tilt_Angle.xlsx', index=False)

# Save the updated DataFrame to a new Json file
df_latitude.to_json('/content/drive/MyDrive/2. Ted Nash/All codes /Code for finding the angle value from the latitude value/JSON file Testing/output_Tilt_Angle1.json')


# Specify the tilt columns you want to plot
tilt_columns = [col for col in df_latitude.columns if col.startswith('Cutomer')]

# Plotting
plt.figure(figsize=(12, 8))

# Bar chart for initial values (first row)
initial_values = df_latitude.iloc[0]
plt.bar(initial_values.index, initial_values, color='blue', alpha=0.7, label='Initial Values')

# Set labels and title
plt.xlabel('Parameter')
plt.ylabel('Value')
plt.title('Initial Values for Latitude Parameters')

# Add legend
plt.legend()

# Show the plot
plt.show()

# Specify the tilt columns you want to plot
tilt_columns = [col for col in df_latitude.columns if col.startswith('Tilt')]

# Plotting
plt.figure(figsize=(12, 8))

for column in tilt_columns:
    latitude_number = column.split("_")[1]
    plt.plot(df_delta['Day'], df_latitude[column], label=f'Customer {latitude_number} Tilt angle (365 days)')



# Set labels and title
plt.xlabel('Day of the Year')
plt.ylabel('Tilt Angle')
plt.title('Tilt Angle for Each Latitude Over 365 Days')


# Add legend
plt.legend()

# Save the plot as an image file
plt.savefig('/content/drive/MyDrive/2. Ted Nash/All codes /Code for finding the angle value from the latitude value/JSON file Testing/Tilt_Angle_Graph_Output.png')

# Show the plot
plt.show()
```
